Check for outliers using bounds.

```python
lower_bound = 0.145
upper_bound = 0.97
res = df.WT.quantile([lower_bound, upper_bound])
print(res)
```

Return an array showing true if a value is an outlier and false if it is not.

```python
true_index = (res.loc[lower_bound] < df.WT.values) & (df.WT.values < res.loc[upper_bound])
false_index = ~true_index
print(false_index)
```

<b> Using Z score for outlier detection </b>

Z score = (mean / std. deviation)

Formula for Z score = (Observation - Mean) / Standard Deviation

```python
m = np.mean(df.WT)
s = np.std(df.WT)

print("Mean:", m, "& Standard Deviation:", s)
```

```python
outliers=[]
def detect_outliers(data):
    threshold=3
    for i in data:
        z_score = (i-m)/s
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers

outlier_pt = detect_outliers(df.WT)
print("Outliers:", outlier_pt)

# Replace values > 100 with 0
df.loc[df.WT > 100, 'WT'] = np.nan
df.fillna(0, inplace=True)
```

<b> Transformation </b>
(Change the scale of a variable using MinMaxScaler)

```python
marks = df[['WT', 'DSBDA', 'AI', 'CC']].copy()

scaler = MinMaxScaler()

df_scaled = scaler.fit_transform(marks.to_numpy())
df_scaled = pd.DataFrame(df_scaled, columns=['WT', 'DSBDA', 'CC', 'AI'])

print("Scaled Dataset using MinMaxScaler:")
df_scaled.head()
```

<b>Transformation</b>
(Change the scale of variable using Pandas and Numpy)

```python
# Min-Max Normalization
df_norm = (marks-marks.min())/(marks.max()-marks.min())

df_norm.insert(loc = 0, column='First_Name', value=df.First_Name)
df_norm.insert(loc = 1, column='Last_Name', value=df.Last_Name)

print("Scaled dataset using Pandas: ")
df_norm.head()
```
